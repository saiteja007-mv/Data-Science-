```python
import pandas as pd
import numpy as np
data = pd.read_csv("Pfizer_1.csv")
```

```python
data.head()
```

```python
data_melt = pd.melt(data, id_vars=['Date','Drug_Name','Parameter'],var_name='Time',value_name='Reading')
```

```python
data_melt.head()
```

```python
data_tidy = data_melt.pivot(index=['Date','Drug_Name','Time'],columns='Parameter',values='Reading').reset_index()
```

```python
data_tidy.head()
```

# Groupping using drug name and apply function

```python
def temp_mean(x):
    x['Average temparature'] = x['Temperature'].mean()
    return x

data_tidy = data_tidy.groupby('Drug_Name').apply(temp_mean)
```

```python
data_tidy
```

```python
data_tidy[:20]
```

# Filling the null values of Temparature and pressure using mean

```python
data_tidy.Temperature.fillna(data_tidy.Temperature.mean(),inplace=True)
```

```python
data_tidy.Pressure.fillna(data_tidy.Pressure.mean(),inplace=True)
```

```python
data_tidy.isna().sum()
```

# Binning the data using cut function in pandas

```python
data_tidy.Temperature.min()
```

```python
data_tidy.Temperature.max()
```

```python
data_tidy.Pressure.min()
```

```python
data_tidy.Pressure.max()
```

```python
temp_points = [5,20,35,50,65]
temp_lables = ['low','medium','high','very_high']
data_tidy['Temparature category']= pd.cut(data_tidy.Temperature,bins=temp_points,labels=temp_lables)
```

```python
data_tidy
```

```python
press_points = [5,15,16,25]
press_lables = ['Below_average','Average','Above_average']
data_tidy['Pressure category']= pd.cut(data_tidy.Pressure,bins=press_points,labels=press_lables)
```

```python
data_tidy['Pressure category'].value_counts()
```

```python
data_tidy['Temparature category'].value_counts()
```

# Getting year from Date column

```python
def get_year(x):
    return x[2]
data_tidy['Year'] = data_tidy['Date'].str.split('-').apply(get_year)
```

```python
data_tidy['Time stamp'] = data_tidy['Date'] + ' ' + data_tidy['Time']
```
