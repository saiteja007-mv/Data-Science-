```python
import numpy as np
import pandas as pd
movies = pd.read_csv("movies.csv",index_col=0)
directors = pd.read_csv("directors.csv",index_col=0)
data = pd.merge(movies,directors,left_on="director_id",right_on='id',how='left')
data.drop('id_y',axis=1,inplace=True)
data.rename({"id_x":"movies_id"},axis=1,inplace=True)
data
```

# How the multi indexing works

```python
data_agg = data.groupby('director_name')[['title','year']].aggregate({'title':'count','year':['min','max']})
data_agg
```

```python
data.columns
```

```python
data_agg.columns
```

# Changing the Multi index to Single index

```python
data_agg.columns = ['_'.join(tuple) for tuple in data_agg.columns]
```

```python
data_agg
```

# To convert the fat data into thin data
* Pandas has a function named melt

```python
data_melt = pd.melt(data, id_vars=['Date','Drug_Name','Parameter'],var_name='Time',value_name='Reading')
```

```python
data_melt.shape
```

```python
data_melt.info()
```

# Change thin data into fat data 
* Pandas has a function called Pivot

```python
data_melt.pivot(index=['Date','Drug_Name','Parameter'],columns = 'Time',values = 'Reading').reset_index()
```

# How to deal with NULL values

### Check whether there are null values

```python
data.isnull().sum(axis=1)
```

### Dropping the null values

```python
data.dropna(axis=0)
```

### Filling the null values with 0

```python
data.fillna(0)
```

### Fill the NULL Values with Average

```python
data['2:30:00'].fillna(data['2:30:00'].mean())
```
