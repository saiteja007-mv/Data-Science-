# Add a new row

```python
df = pd.read_csv("mckinsey (1).csv")
```

```python
df.head()
```

#### Creating a dict of new values

```python
new_row = {'country':'India','year':1988,'population':94343233,'continent':'Asia','life_exp':89.99,'gdp_cap':679.90}
```

```python
new_row
```

```python
df.loc[1704] = new_row
```

```python
df.tail(3)
```

```python
df
```

# Update the single cell value

#### data_frame.at[row_number, column_name] = value_to_update

```python
df.at[1703,'population'] = 29601212.324530516
```

```python
n = int(input('Enter how many extra row you want to add:'))
for i in range(0,n):
    df.loc[len(df.index)+i] = new_row
    print(len(df.index)+i)
df.reset_index(inplace=True)
```
