#### 1. Find the profit generated by each movie. (use Apply function)

```python
def profit(row):
    return row.revenue - row.budget

movies['profit'] = movies.apply(profit,axis=1)
```

```python
movies = pd.read_csv("movies.csv",index_col=0)
directors = pd.read_csv("directors.csv",index_col=0)
data = pd.merge(movies,directors,left_on="director_id",right_on='id',how='left')
data.drop('id_y',axis=1,inplace=True)
data.rename({"id_x":"movies_id"},axis=1,inplace=True)
data
```

```python
fizer = pd.read_csv("Pfizer_1.csv")
```

#### 1. Find the list of movies of all the directors whose movie average is less than their over all movies average 

```python
def average(x):
    return x.loc[x['vote_average'] < x['vote_average'].mean()]
result = data.groupby('director_name').apply(average)
```

```python
result
```

#### 2. Find out which director directed highest number of movies in a single year

```python
movies_per_year = data.groupby(['director_name','year'])['movies_id'].count().reset_index(name='movies_count')
```

```python
max_movies = movies_per_year['movies_count'].max()
```

```python
director_highest_noofmovies = movies_per_year[movies_per_year['movies_count'] == max_movies]
director_highest_noofmovies
```

#### 3. In the movies table merge the Year and Month columns to make it a single column to display it as (Month-Year, Ex: Dec-2023)

```python
data['Month_year'] = data['month'] + '-' + data['year'].astype('str')
```

#### 4. Group all the movies based on vote average and display those movies between 7.0 to 9.0 in descending order 

```python
filtered = data.loc[(data.vote_average>=7.0) & (data.vote_average<=9.0)]

groupped = filtered.groupby('vote_average')
```

```python
groupped.groups
```

```python
for vote_avg, group in groupped:
    print(group[['title','vote_average']].sort_values(by='vote_average',ascending=False))
```
