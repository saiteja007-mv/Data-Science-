# Correlation 
## If two values are correlated
### Positively Correlated - If one value increase other value also increase
### Negative Correlated - If one increase other descrease vice versa
#### For example - If you take more calories the weight will also increase So Calories and Weight are positively Correlated

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("final_vg.csv")
```

```python
data.head()
```

# Suppose we have find relation between two data points

# Scatter Plot using Seaborn

```python
sns.scatterplot(data,x='Rank',y='Global_Sales')
```

* In this scatter plot we can observe that when there is higher sales the rank is less and vice-versa
* So it is negatively correlated

### Same can be visualized using line plot

# Line plot using Seabornm

```python
sns.lineplot(data,x='Rank',y='Global_Sales')
```

* But it is not that effective So we consider scatter plot

#### Upto now we looked into continous - continous data

# Categorical - Categorical data

```python
data.head(3)
```

```python
top3_pub = data['Publisher'].value_counts().index[:3]
top3_gen = data['Genre'].value_counts().index[:3]
top3_plat = data['Platform'].value_counts().index[:3]
```

```python
top3_pub
```

```python
top3_gen
```

```python
top3_plat
```

```python
top3_data = data.loc[(data['Publisher'].isin(top3_pub)) & (data['Genre'].isin(top3_gen)) & (data['Platform'].isin(top3_plat))]
```

```python
top3_data
```

# Dodged Bar chart

```python
sns.countplot(x='Publisher',data=top3_data,hue='Platform')
```

# Multivariate data

# Heatmap - It shows the correlation between numerical columns

```python
top3_multi = top3_data.drop(['Name','Unnamed: 0','Platform','Genre','Publisher'],axis=1)
```

```python
sns.heatmap(top3_multi.corr(),cmap='Blues',annot=True)
```
