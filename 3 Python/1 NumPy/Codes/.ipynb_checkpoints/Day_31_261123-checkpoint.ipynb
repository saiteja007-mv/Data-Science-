# Changing Image colors

```python
img_r = img.copy()
img_r[:,:,(1,2)] = 0
plt.imshow(img_r)
```

```python
img_g = img.copy()
img_g[:,:,(0,2)] = 0
plt.imshow(img_g)
```

```python
img_b = img.copy()
img_b[:,:,(0,1)] = 0
plt.imshow(img_b)
```

```python
img_m = img.copy()
img_m[:500,:500,(1,2)] = 0
plt.imshow(img_m)
```
